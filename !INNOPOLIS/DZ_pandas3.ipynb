```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
bicycle_data = pd.read_csv('hour.csv', index_col='instant')
bicycle_data
```

```python
print("Первые 5 строк датасета:")
print(bicycle_data.head())
print("\nНазвания столбцов в датасете:")
print(bicycle_data.columns)
print("\nКоличество пропущенных значений по столбцам:")
print(bicycle_data.isnull().sum())
```

```python
h_day_N=0
w_day_N=0
for index, row in bicycle_data.iterrows():
    if row['holiday']:
        h_day_N+=row['cnt']
    else:
        w_day_N+=row['cnt']
print(f'В выходные было арендовано {h_day_N} велосипедов, в будни — {w_day_N}.')
```

```python
new_bicycle_data = pd.get_dummies(bicycle_data,columns=['season','yr','mnth','weekday','weathersit'], dtype=int)  # Преобразуем категориальные переменные
```

```python
avg_by_mnth = bicycle_data.groupby('mnth')['cnt'].mean().apply(lambda x: round(x, 2))
print("Среднее количество арендованных велов по месяцам:")
avg_by_mnth
```

```python
# пропущенных значений у нас нет, как мы выяснили, но предположим, что они есть:
bicycle_data = bicycle_data.fillna(0)
print("\nПример данных после заполнения пропусков:")
bicycle_data
```

```python
sns.barplot(x='hr', y='cnt', data=bicycle_data[['hr', 'cnt']].groupby(['hr'], as_index=True).mean(), errorbar=None,label='Mean count per hour')
```
