```python
class Bk:
    def __init__(self, name, author,year: int,cnt:int):
        self.name = name
        self.author = author
        self.year=year
        self.__cnt = cnt if cnt else 0
        
    def __str__(self):
        return f'Книга «{self.name}», {self.author}, {self.year}г. изд. в количестве {self.__cnt} шт.'
    
    @property
    def check(self):
        return True if self.__cnt >0 else False

    def cntUp(self):
        self.__cnt+=1
        return self.__cnt

    def cntDown(self):
        if self.__cnt >0:
            self.__cnt-=1
        else:
            print(f'Книга {self.name} закончилась!')
        return self.__cnt

class Rdr:
    name=""
    books=[]
    def __init__(self, name):
        self.name = name
        self.books=[]
        
    def __str__(self):
        return f'Читатель {self.name},  книг на нём {len(self.books) if self.books else 0}'

    def bksLst(self):
        print(f'{self.name} читает {len(self.books) if self.books else 0} книг:')
        for elem in self.books: # Итерация 
            print(elem)
    
    def getBk(self,book):
        if book.check:
            self.books.append(book)
            book.cntDown()
            return True
        else:
            print(f'Не дают такую книгу(')
            return False

    def returnBk(self,book):
        if book in self.books:
            self.books.remove(book)
            book.cntUp()
            return True
        else:
            print(f'{self.name} такого не брал!')
            return False

class Lbr:
    name=""
    books=[]
    readers=[]
    def __init__(self, name):
        self.name = name
        self.books=[]
        self.readers=[]
        
    def __str__(self):
        return f'Библиотека {self.name},  книг в ней {len(self.books) if self.books else 0}, читателей зарегистрировано  {len(self.readers) if self.readers else 0}.'

    def bksLst(self):
        print(f'В библиотеке {self.name} имеется {len(self.books) if self.books else 0} книг:')
        for elem in self.books: # Итерация 
            print('     ',self.books.index(elem)+1,'. ', elem)

    def rdrsLst(self):
        print(f'В библиотеке {self.name} зарегистрировано {len(self.readers) if self.readers else 0} читателей:')
        for relem in self.readers: # Итерация 
            print('     ',self.readers.index(relem)+1,'. ', relem)
            for belem in relem.books:
                print('            * ', belem.name)

    def addBk(self,book): #добавление новой книги с проверкой - лишнего не добавит
        if not (self.search(book.name) or self.search(book.author)):
            self.books.append(book)
            return True
        else:
            print(f'Такая книга уже есть, хватит!')
            return False

    def delBk(self,book): #удаление книги с проверкой
        if (self.search(book.name) and self.search(book.author)):
            self.books.remove(self.search(book.name))
            return True
        else:
            print(f'Таких книг у нас и не было!')
            return False

    def addRdr(self,reader): #добавление читателя. без проверки - читают все!!
        self.readers.append(reader)
        return True

    def delRdr (self,reader_name): #удаление читателя - без проверки! если таких несколько, удалит первого подходящего. если такого нет - просто вернем False
        for reader in self.readers:
            if reader.name.lower().find(reader_name.lower())>=0:
                self.readers.remove(reader)
                return True
        return False

    def search(self, what:str): # ищет первую (!) книгу по вхождению строки в название или в автора. поиск регистронезависимый.
        for book in self.books:
            if book.name.lower().find(what.lower())>=0:
                return book
            elif book.author.lower().find(what.lower())>=0:
                return book
        return False

    def rdrSearch(self, who:str): # ищет первого (!) читателя по вхождению строки в имя. поиск регистронезависимый.
        for reader in self.readers:
            if reader.name.lower().find(who.lower())>=0:
                return reader
        return False
```

```python
###main#################################################################################
N=1
state='' #выбор пункта меню
print(f'Добро пожаловать в нашу библиотеку!')
while state !='0':
    try:
        print('''___________________________________________________________________________________________________________________________________________________________________________________________________
| 1-зарегистрироваться | 2-поиск книги | 3-взять книгу | 4-сдать книгу | 5-подарить книгу библиотеке | 6-удалить книги из библиотеки | 7-а можно все посмотреть? | 8-исключить читателя | 0-выйти |
‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾''')
        state=input()[:1]
        match state:
            case '1': #регистрация читателя
                new_reader_name =str(input('Представьтесь, пожалуйста, будущий читатель:'))
                new_reader=Rdr(new_reader_name)
                NskLib.addRdr(new_reader)
                NskLib.rdrsLst()
            case '2': # ищем книгу по строке. выводим первую найденную
                search_book = NskLib.search(str(input('Что ищем?')))
                if search_book:
                    print(search_book)
                else:
                    print('Нет у нас такой книги!')
            case '3': #взять книгу для конкретного читателя
                get_book = NskLib.search(str(input('Что ищем?')))
                if get_book:
                    get_reader=NskLib.rdrSearch(str(input('На кого берем?')))
                    if get_reader:
                        if get_reader.getBk(get_book):
                            print(f'Готово! {get_reader}, он взял «{get_book.name}»! \nНе забудь вернуть, {get_reader.name}!')
                    else:
                        print('Нет у нас такого читателя!')
                else:
                    print('Нет у нас такой книги!')
            case '4': #сдать книгу от конкретного читателя
                put_reader=NskLib.rdrSearch(str(input('Кто возвращать будет?')))
                if put_reader:
                    return_book = NskLib.search(str(input('Что возвращаем?')))
                    if return_book:
                        if put_reader.returnBk(return_book):
                            print(f'Готово! {put_reader}, он вернул «{return_book.name}»! \nМолодец, {put_reader.name}!')
                    else:
                        print('Нет у нас такой книги!')    
                else:
                    print('Нет у нас такого читателя!')
            case '5': #добавить книгу в конкретном количестве экземпляров в библиотеку
                add_book_name =str(input('Книгу принесли? Как называется? — '))
                add_book_author =str(input('Кто написал? — '))
                add_book_year =int(input('В каком году? — '))
                add_book_cnt =int(input('А сколько штук принесли? — '))
                add_book=Bk(add_book_name,add_book_author,add_book_year,add_book_cnt)
                NskLib.addBk(add_book)
                print(f'Вот спасибо, хорошо, положите на комод!')
            case '6': #удаление всех экземпляров книги из библиотеки
                del_book_name =str(input('Что будем удалять? — '))
                del_book=NskLib.search(del_book_name)
                if del_book:
                        NskLib.delBk(del_book)
                else:
                    print('Нет у нас такой книги!')
            case '7': #вывод всего списка книг библиотеки, чисто на посмотреть
                NskLib.bksLst()
            case '8': #удаление читателя по имени. если таких несколько - первого найденного
                ex_reader_name =str(input('Кого будем исключать? — '))
                NskLib.delRdr(ex_reader_name)
                NskLib.rdrsLst()
            case '0':
                pass
            case _:
                raise Exception
    except Exception:
        if N<3:
            print('Что-то пошло не так! Попробуйте еще раз.')
            N += 1
        else:
            print('Превышено количество неудачных попыток!')
            break  
```
