```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv('2016.csv')

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = df['Happiness Score']
y = df['Economy (GDP per Capita)']
z = df['Health (Life Expectancy)']

ax.set_xlabel("")
ax.set_ylabel("")
ax.set_zlabel("")

ax.scatter(x, y, z)
```

```python
dataframe = pd.read_csv("forestfires.csv")
print(dataframe)
dataframe.head()
```

```python
dataframe.isnull().values.any()
```

```python
FFMC = dataframe["FFMC"]
DMC = dataframe["DMC"]
DC = dataframe["DC"]
RH = dataframe["RH"]
ISI = dataframe["ISI"]
temp = dataframe["temp"]
```

```python
sns.boxplot(x=DMC)
```

```python
sns.boxplot(x=dataframe["day"], y=DMC)
```

```python
df = pd.DataFrame(data=dataframe, columns=["FFMC", "DMC", "DC", "ISI"])
sns.boxplot(x="variable", y="value", data=pd.melt(df))
```

```python
colors = ['#78C850', '#F08030', '#6890F0','#F8D030', '#F85888', '#705898', '#98D8D8']
sns.boxplot(x=DMC, y=dataframe["day"], palette=colors)
```

```python
df = pd.DataFrame(data=dataframe, columns=["FFMC", "DMC", "DC", "ISI"])
boxplot = sns.boxplot(x="variable", y="value", data=pd.melt(df))
boxplot.axes.set_title("    ", fontsize=16)
boxplot.set_xlabel("", fontsize=14)
boxplot.set_ylabel("", fontsize=14)
```

```python
df = pd.DataFrame(data=dataframe, columns=["FFMC", "DMC", "DC", "ISI"])
boxplot = sns.boxplot(x="variable", y="value", data=pd.melt(df), order=["DC", "DMC", "FFMC", "ISI"])
boxplot.axes.set_title("    ", fontsize=16)
boxplot.set_xlabel("", fontsize=14)
boxplot.set_ylabel("", fontsize=14)
```
