```python
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://book24.ru/')
bs = BeautifulSoup(res.text, 'lxml')
print(bs.prettify())
```

```python
head_tag = bs.head
title_tag = head_tag.contents[1]
for child in title_tag.children:
    print(child)
```

```python
bs.select("head > title")
```

```python
authors = bs.select('[class ~= author-list__item]')
for author in authors[:10]:
    print(author.string)
```

```python
authors = bs.select('[class ~= author-list__item]')
for author in authors[:10]:
    print(author.get_text())
```

```python
title_tag = bs.title
print(title_tag)
print(title_tag.parent)
```
