```python
#7
pd.to_numeric(Gdata['Rating'], downcast="integer",errors='coerce')
Gdata[Gdata['Rating']>=4.9].head(10)
```

```python
#8 работает некорректно - по идее, errors='coerce' у функции to_numeric долждно в случае ошибок конвертации установить значение в NaN, но оно радостно пишет строки в якобы numeric-значения, а потом еще и сортирует их...
pd.to_numeric(Gdata['Installs'], errors='coerce')
Gdata.sort_values(by='Installs', ascending=False,na_position='first').head(5)
```

```python
#9
def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{absolute:d} шт."
    
Gdata['I'] = Gdata.index
pdt=Gdata.pivot_table(values='I', columns='Category', aggfunc='count')
plt.pie(pdt.loc['I'],labels=pdt.columns,autopct=lambda pct: func(pct, pdt.loc['I']),radius=2, pctdistance=0.8)
plt.show()
```
