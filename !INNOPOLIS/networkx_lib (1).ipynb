```python
import networkx as nx
```

```python
G = nx.Graph()
print(G)
```

```python
G.add_node(1)
print(G)
```

```python
G.add_nodes_from([2, 3])
print(G)
nx.draw(G)
```

```python
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])
print(G)
nx.draw(G)
```

```python
G.nodes()
G.edges()
```

```python
nx.draw(G)
```

```python
G.add_edge(1, 2)
nx.draw(G)
```

```python
e = (2, 3)
G.add_edge(*e) 
G.add_edges_from([(1, 3), (3, 4)])
nx.draw(G)
```

```python
H = nx.path_graph(10)
print(H)
```

```python
nx.draw(H)
```

```python
G.add_nodes_from(H)
print(G)
nx.draw(G,with_labels=True)
```

```python
G.remove_node(2)
G.remove_node('m')
nx.draw(G,with_labels=True)
```

```python
G.clear()
print(G)
```

```python
G.add_edges_from([(1, 2), (1, 3)])
G.add_node(1)
G.add_edge(1, 2)
G.add_node("spam")
G.add_nodes_from("spam")
G.add_edge(3, 'm')
print(G)
nx.draw(G,with_labels=True)
```

```python
edgelist = [(0, 1), (1, 2), (2, 3)]
```

```python
H = nx.Graph(edgelist)
list(H.edges())
```

```python
adjacency_dict = {0: (1, 2), 1: (0, 2), 2: (0, 1)}
H = nx.Graph(adjacency_dict) 
list(H.edges())
```

```python
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1, 2, 0.5), (3, 1, 0.75)])
nx.draw(DG,with_labels=True)
```

```python
import matplotlib.pyplot as plt
nx.draw(DG,with_labels=True)
plt.savefig("example.png")
```
