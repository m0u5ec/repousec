```python
# Загружаем датасет Titanic
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# 1. Обнаружение пропущенных значений
missing_values = titanic.isnull().sum()
print("Пропущенные значения по столбцам:")
print(missing_values)

# 2. Удаление строк с пропусками в столбце 'Cabin'
titanic_cleaned = titanic.dropna(subset=['Cabin'])
print("\nЧисло строк после удаления строк с пропусками в 'Cabin':", len(titanic_cleaned))

# 3. Заполнение пропусков в столбце 'Age' средним значением
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
print("\nПример данных после заполнения пропусков в 'Age':")
titanic[['Name', 'Age']].head()
```

```python
# Загружаем датасеты Titanic и Weather
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
weather = pd.DataFrame({
    'Date': ['1912-04-10', '1912-04-11'],
    'Temperature': [55, 58],
    'Precipitation': [0, 0]
})

# Преобразуем столбец Date в тип datetime
weather['Date'] = pd.to_datetime(weather['Date'])

# Присваиваем дату всем пассажирам (предполагаем, что они все были на борту 10 апреля 1912 года)
titanic['Date'] = pd.to_datetime('1912-04-10')

# Объединяем датасеты по дате
combined_data = pd.merge(titanic, weather, on='Date')

# Итерация по объединенным данным с использованием iterrows()
print("Итерация с использованием iterrows():")
for index, row in combined_data.iterrows():
    print(f"Пассажир: {row['Name']}, Температура: {row['Temperature']}°F, Выживание: {row['Survived']}")

# Итерация по объединенным данным с использованием itertuples()
print("\nИтерация с использованием itertuples():")
for row in combined_data.itertuples(index=False):
    print(f"Пассажир: {row.Name}, Температура: {row.Temperature}°F, Выживание: {row.Survived}")
```

```python
# Задаем путь к файлу
file_path = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv'

# Чтение данных по частям
chunks = pd.read_csv(file_path, chunksize=100000)

# Инициализация пустого списка для хранения отфильтрованных данных
filtered_data = []

# Обработка каждого чанка
for chunk in chunks:
    # Фильтруем данные за определенный период (например, с 2022 года)
    filtered_chunk = chunk[chunk['date'] >= '2022-01-01']
    filtered_data.append(filtered_chunk)

# Объединение всех отфильтрованных данных в один DataFrame
result = pd.concat(filtered_data)

# Выводим первые несколько строк результата
print("Отфильтрованные данные:")
print(result.head())
```

```python
# Загружаем датасет Titanic
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Создание нового столбца с категорией пассажиров
def age_category(age):
    if age < 18:
        return 'Молодой'
    else:
        return 'Взрослый'

titanic['AgeCategory'] = titanic['Age'].apply(age_category)

# Выводим первые несколько строк с новым столбцом
print("Данные с новым столбцом AgeCategory:")
print(titanic[['Name', 'Age', 'AgeCategory']].head())

# Расчет среднего возраста пассажиров в зависимости от класса
average_age_by_class = titanic.groupby('Pclass')['Age'].mean().apply(lambda x: round(x, 2))
print("\nСредний возраст пассажиров в зависимости от класса:")
average_age_by_class
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Загрузка локального датасета о заболеваниях сердца
heart_data = pd.read_csv('heart.csv')

# Проверка названий столбцов
print("Названия столбцов в датасете:")
print(heart_data.columns)
print("\nОписание датасета:")
print(heart_data.describe())
# Проверка на пропущенные значения
print("\nПропущенные значения:")
print(heart_data.isnull().sum())

# 1. Исследование и очистка данных
print("\nПервые 5 строк датасета:")
print(heart_data.head())

# 2. Преобразование категориальных переменных
heart_data = pd.get_dummies(heart_data, drop_first=True)  # Преобразуем категориальные переменные
print("\nЕще раз первые 5 строк датасета после get_dummies:")
print(heart_data.head())

# 3. Анализ корреляции
plt.figure(figsize=(10, 8))
correlation_matrix = heart_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Матрица корреляции')
plt.show()

# 4. Подготовка данных для модели
# Замените 'target' на правильное имя столбца, если оно отличается
target_column = 'target'  # Здесь укажите правильное имя целевой переменной, если оно отличается
if target_column not in heart_data.columns:
    print(f"Целевая переменная '{target_column}' не найдена. Пожалуйста, проверьте названия столбцов.")
else:
    X = heart_data.drop(target_column, axis=1)  # Все столбцы, кроме 'target'
    y = heart_data[target_column]  # Столбец 'target' как целевая переменная

    # Разделение на обучающую и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Построение модели
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)

    # Предсказание и оценка модели
    y_pred = model.predict(X_test)
    print("\nОтчет о классификации:")
    print(classification_report(y_test, y_pred))
    print("Матрица путаницы:")
    print(confusion_matrix(y_test, y_pred))
```
