```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_s=np.arange(0,11,1)
Y_s= [x**2 for x in X_s]
sns.lineplot(x=X_s,y=Y_s,label='X²')
plt.axis([0,11,0,110])
plt.title('График Y=X²')
plt.xlabel('Значения X')
plt.ylabel('Значения X²')
plt.legend(loc=4)
```

```python
X_s=np.arange(0,100,1)
Y_s=np.random.randn(100)
dataframe=pd.DataFrame({'A':X_s,'B':Y_s})
sns.color_palette("light:#5A9", as_cmap=True)
sns.histplot(data=dataframe,x='B', bins=20, label='RandNormal',color='g')
plt.title('Random Normal')
plt.xlabel('Random X')
plt.ylabel('Count')
plt.legend(loc=4)
```

```python
X_s=['A', 'B', 'C', 'D']
Y_s= [40, 30, 20, 10]
plt.pie(Y_s,labels=X_s,colors=['g','m','k','0.8'],explode = [00, 0, 0.1,0])
plt.show()
```
