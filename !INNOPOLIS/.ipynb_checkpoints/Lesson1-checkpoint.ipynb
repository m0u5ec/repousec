# LaTeX

LaTeX можно использовать в Jupyter Notebook для написания математических формул и уравнений. Вот несколько примеров с объяснениями

## 1. Простое уравнение

Чтобы написать простое уравнение, используйте символы $ перед и после LaTeX-кода

$y = mx + b$


## 2. Квадратное уравнение

$ax^2 + bx + c = 0$


## 3. Формула с дробью

$\frac{a}{b}$


## 4. Сумма

$\sum_{i=1}^{n} i = \frac{n(n+1)}{2}$


## 5. Интеграл

$\int_a^b f(x) \, dx$


## 6. Матрица

**Текст, выделенный полужирным шрифтом**$\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}$


## 7. Степени и индексы

$e^{i\pi} + 1 = 0$


## 8. Корни

$\sqrt{a^2 + b^2}$


## 9. Векторы и наборы

$\vec{v} = \begin{pmatrix} x \\ y \\ z \end{pmatrix}$


## 10. Комбинация символов

$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$


# Магические команды

Магические команды в Jupyter Notebook представляют собой специальные команды, которые выполняются с помощью % или %% и служат для расширения возможностей взаимодействия с окружением Jupyter. Эти команды можно использовать для выполнения кода, управления окружением, анализа данных, работы с файлами и многого другого

## 1. Линейные магические команды (одинарный %)

%pwd: Показать текущий рабочий каталог.

````python
%pwd
````

%cd: Изменить текущий рабочий каталог

%cd /path/to/directory


%ls: Показать содержимое текущего каталога

````python
%ls
````

%who: Показать список всех переменных в пространстве имён

````python
a = 10
%who
````

%whos: Показать подробную информацию обо всех переменных

````python
%whos
````

%reset: Сбросить (удалить) все переменные и очистить пространство имён

````python
%reset
````

%time: Измерить время выполнения одной строки кода

````python
import time

def some_function():
    for i in range(5):
        time.sleep(1)

%time some_function()
````

%timeit: Многократно выполнить одну строку кода и замерить среднее время выполнения.

````python
import time

def some_function():
    for i in range(5):
        time.sleep(1)
%timeit some_function()
````

%run: Выполнить Python-скрипт в рамках текущего сеанса.

``%run script.py``


%history: Показать историю выполнения команд.

````python
%history
````

%matplotlib: Включить встроенную поддержку Matplotlib (например, для отображения графиков в ноутбуке).

````python
%matplotlib inline
````

%env: Показать или установить переменные окружения.

````python
%env MY_VARIABLE=value
````

%load: Загрузить код из файла или URL в текущую ячейку

%load script.py`


%store: Сохранить переменные в файл, чтобы использовать их в последующих сеансах

````python
%store a
````

## Ячеечные магические команды (двойной %%)

%%time: Измерить время выполнения кода в ячейке.

%%timeit: Многократно выполнить код в ячейке и замерить среднее время выполнения.

%%writefile: Записать содержимое ячейки в файл.

%%writefile script.py


%%capture: Захватить вывод команд в ячейке (stdout, stderr) и сохранить его в переменную.

%%bash: Выполнить код в ячейке как Bash-скрипт.

%%perl: Выполнить код в ячейке как Perl-скрипт

%%ruby: Выполнить код в ячейке как Ruby-скрипт.

%%html: Отобразить содержимое ячейки как HTML


````python
%%html
<h1>Hello, World!</h1>
````

%%javascript: Выполнить JavaScript-код.

````python
%%javascript
alert("Hello, World!");
````

%%latex: Отобразить содержимое ячейки как LaTeX.

````python
%%latex
\LaTeX{} is great!
````

%%markdown: Отобразить содержимое ячейки как Markdown.

````python
%%markdown
# Hello, World!
````

%%python: Явно указать выполнение Python-кода в ячейке (полезно, если используется другое ядро).

````python
%%python
print("Hello, World!")
````

## Особые магические команды

%magic: Показать справочную информацию по магическим командам.

````python
%magic
````

%debug: Войти в отладчик по завершении исключения.

````python
%debug
````

%pdb: Включить автоматический вход в отладчик при возникновении ошибки.

````python
%pdb on
````

%memit: Измерить объём памяти, используемый объектом (необходимо установить memory_profiler).

%mprun: Выполнить код с профилированием памяти (необходимо установить memory_profiler).
