```python
import pandas as pd
import seaborn as sns
Tdata = pd.read_csv('titanic_train.csv', index_col='PassengerId')
isAge = pd.notna(Tdata['Age'])
Tdata["isAge"] = isAge
Tdata.head(15)
```

```python
Tdata.groupby('Sex').isAge.value_counts()
```

```python
Tdata[['Sex', 'isAge']].groupby(['Sex'], as_index=True).sum()
```

```python
sns.barplot(x='Sex', y='isAge', data=Tdata[['Sex', 'isAge']].groupby(['Sex'], as_index=True).sum(), errorbar=None,label='Known age count')
```
