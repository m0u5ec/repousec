```python
###Реализовать консольную программу для простых операций с изображениями с помощью модуля OpenCV
###(обрезка, поворот, масштабирование, отзеркаливание, размытие, увеличение резкости). 
###Реализовать полноценное консольное меню с возможностью считывания файла, сохранения, выхода из программы. 
###В программе реализовать проверку на возможные исключения. 
###Также в программе должна присутствовать возможностью загрузки более одного файла за раз и применения к ним одной выбранной операции, с последующем сохранением. 
###При сохранении файла, от пользователя требуется указать как путь, так и имя файла. 
###К имени файла при сохранении должно добавляться текущая дата и время (image-27.11-10.12.jpg)
###Постараться придерживаться принципа: 1 функция - 1 действие. 
```

```python
import os
import cv2
import cv2_ext
import numpy as np
from datetime import datetime, date, time

###2###################################################################################
def save_now(file_path, file_obj) -> str: #сохраняет объект в файл в текущей категории, сформировав имя с временной меткой
    try:
        full_path=os.getcwd() + '\\' + file_path.split('.')[0] + '_' + datetime.now().strftime('%Y-%m-%d-%H-%M-%S.') + file_path.split('.')[1]
        cv2_ext.imwrite(full_path, file_obj)
        return f'Успешно сохранён файл {full_path}'
    except Exception:
        return f'Что-то пошло не так при сохранении файла {full_path}'
###4####################################################################################
def cut_now(file_path,file_obj): #отсчитывая от центра изображения обрезает картинку до новой ширины+высоты. если ввели ширину/высоту больше текущей - ошибка и на выход
    try:
        wide = float(input(f'Введите новую ширину для файла {file_path}:'))
        if wide>=file_obj.shape[1]:
            raise ValueError
        height = float(input(f'Введите новую высоту для файла {file_path}:'))
        if height>=file_obj.shape[0]:
            raise ValueError
        cut_file_obj = file_obj[int((file_obj.shape[1]-wide) / 2):int((file_obj.shape[1]-wide) / 2 + wide), int((file_obj.shape[0]-height) / 2):int((file_obj.shape[0]-height) / 2 + height)]
        print(f'Файл {file_path} успешно обрезан до размеров {wide} X {height}')
        return cut_file_obj
    except ValueError:
        print(f'Размеры должны быть в числовом виде и не должны превышать текущие!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при обрезке файла {file_path}')
        return file_obj
###5####################################################################################
def resize_now(file_path,file_obj): #по введенному коэффициенту масштабирования пересчитывает новые размеры картинки, сохраняя пропорции. коэффициент должен быть положительным, от 0 до 1 - уменьшение, >1 увеличение.
    try:
        koeff = float(input(f'Введите коэффициент масштабирования для файла {file_path}:'))
        if koeff<=0:
            raise ValueError
        new_size=(int(file_obj.shape[1] * koeff), int(file_obj.shape[0] * koeff))
        resized_file_obj =  cv2.resize(file_obj, new_size, interpolation=cv2.INTER_AREA)
        print(f'Файл {file_path} успешно масштабирован до размеров {new_size[0]} X {new_size[1]}')
        return resized_file_obj
    except ValueError:
        print(f'Коэффициент должен быть положительным числом!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при масштабировании файла {file_path}')
        return file_obj    
###6####################################################################################
def rotate_now(file_path,file_obj): #поворачивает картинку 1 - по горизонтали, 2 - по вертикали, любой другой символ - в обоих плоскостях.  
    try:
        print(f'Поворачиваем файл {file_path}:')
        mode = str( input(f'''_______________________________________________________________
| 1-по горизонтали | 2-по вертикали | иное-в обоих плоскостях |
‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾''')[:1])
        match mode:
            case '1':
                rotated_file_obj=cv2.flip(file_obj,1)
            case '2':
                rotated_file_obj=cv2.flip(file_obj,0)
            case _:
                rotated_file_obj=cv2.flip(file_obj,-1)
        print(f'Файл {file_path} успешно повёрнут')
        return rotated_file_obj
    except ValueError:
        print(f'Ерунду какую-то ввели!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при повороте файла {file_path}')
        return file_obj    
###7####################################################################################
def blur_now(file_path,file_obj): #размытие картинки.  
    try:
        blured_file_obj = cv2.GaussianBlur(file_obj, ksize=(11, 11), sigmaX =0, sigmaY=0)    
        print(f'Файл {file_path} успешно размыт')
        return blured_file_obj
    except ValueError:
        print(f'Ерунду какую-то ввели!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при размытии файла {file_path}')
        return file_obj    
###8####################################################################################
def clear_now(file_path,file_obj): #увеличение резкости картинки.  
    try:
        matrix = [[-1, -1, -1],[-1, 9, -1],[-1, -1, -1]]
        sharp_filter = np.array(matrix)
        clear_file_obj =  cv2.filter2D(file_obj, ddepth=-1, kernel=sharp_filter)
        print(f'Файл {file_path} успешно прояснЁн')
        return clear_file_obj
    except ValueError:
        print(f'Ерунду какую-то ввели!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при увеличении резкости файла {file_path}')
        return file_obj    

###9####################################################################################
def depress_now(file_path,file_obj): #Перевод картинки в ЧБ.  
    try:
        depressed_file_obj =  cv2.Sobel(file_obj, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)   
        print(f'Ура( Файл {file_path} теперь весьма депрессивен :(')
        return depressed_file_obj
    except ValueError:
        print(f'Ерунду какую-то ввели!')
        return file_obj
    except Exception:
        print(f'Что-то пошло не так при переводе файла {file_path} в ЧБ')
        return file_obj
###main#################################################################################
N=1
state='' #выбор пункта меню
current_files=[] #список имен файлов (попадают сюда после проверки на существование)
images=[] #список объектов изображений (по индексам соответствуют именам файлов в массиве current_files)
while state !='0':
    try:
        print(len(current_files),' файл(а/ов) в работе: ',*current_files)
        print('''__________________________________________________________________________________________________________________________________
| 1-загрузить | 2-сохранить | 3-показать | 4-обрезать | 5-масштабировать | 6-повернуть | 7-размыть | 8-резкость | 9-ЧБ | 0-выйти |
‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾‾''')
        state=input()[:1]
        match state:
            case '1': #загрузить файлы для работы. если в рабочих массивах уже что-то есть - загружает дополнительрно к ниим. Можно загрузить несколько за раз.
                print('Текущая директория ', os.getcwd())
                for path in input('Введите имена файлов (через запятую):').split(','):
                    if os.path.exists(os.getcwd() + '\\' + path.strip()):
                        current_files.append(path.strip())
                        images.append(cv2_ext.imread(os.getcwd() + '\\' + path.strip()))
                        ##images.append(cv2.imdecode(np.fromfile(os.getcwd() + '\\' + path.strip(), dtype=np.uint8), cv2.IMREAD_UNCHANGED))
                    else:
                        print (f'Файл {path} не найден!')
            case '2': # сохранить все текущие файлы и очистить рабочие списки
                for file in current_files:
                    print(save_now(file,images[current_files.index(file)]))
                current_files.clear()
                images.clear()
            case '3': ##показать — открывает ТЕКУЩИЕ варианты файлов в работе, по очереди, каждое изображение в отдельном окне. почему-то при выносе этого добра в функцию создаваемое окно виснет намертво и ничего не отображает, поэтому оставила тут.
                for file in current_files:
                    cv2.startWindowThread()
                    cv2.namedWindow(file, cv2.WINDOW_AUTOSIZE)
                    cv2.imshow(file,images[current_files.index(file)])
                    cv2.setWindowProperty(file, cv2.WND_PROP_TOPMOST, 1)
                    cv2.waitKey(0)
                    cv2.destroyAllWindows()
            case '4': #обрезать картинки. вынесено в функцию
                for file in current_files:
                   images[current_files.index(file)] = cut_now(file,images[current_files.index(file)])
            case '5': #масштабировать картинки. вынесено в функцию
                for file in current_files:
                   images[current_files.index(file)] = resize_now(file,images[current_files.index(file)])
            case '6': #повернуть картинки. вынесено в функцию
                for file in current_files:
                   images[current_files.index(file)] = rotate_now(file,images[current_files.index(file)])
            case '7': #размыть картинки. вынесено в функцию
                for file in current_files:
                   images[current_files.index(file)] = blur_now(file,images[current_files.index(file)])
            case '8': #"прояснить" картинки. вынесено в функцию
                for file in current_files:
                   images[current_files.index(file)] = clear_now(file,images[current_files.index(file)])
            case '9': #Перевод картинки в ЧБ. вынесено в функцию, НО! почему-то при выносе в функцию работает оооочень странно - как будто применяет блюр+filter2D+ЧБ одновременно! При вызове из основного блока кода работало нормально, но оставила в функции - так депрессивнее)
                for file in current_files:
                   images[current_files.index(file)] = depress_now(file,images[current_files.index(file)])
            case '0':
                pass
            case _:
                raise Exception
    except Exception:
        if N<3:
            print('Что-то пошло не так! Попробуйте еще раз.')
            N += 1
        else:
            print('Превышено количество неудачных попыток!')
            break  
```
