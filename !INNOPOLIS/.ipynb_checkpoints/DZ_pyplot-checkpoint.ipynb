```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_s=np.arange(0,11,1)
Y_s= [x**2 for x in X_s]
plt.axis([0,11,0,110])
plt.title('График Y=X²')
plt.xlabel('Значения X')
plt.ylabel('Значения X²')
plt.grid()
plt.plot(X_s,Y_s,label='X²')
plt.legend(loc=4)
plt.show()
```

```python
X_s=np.arange(0,101,1)
Y_s=np.random.randn(100)
plt.title('Гистограмма')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')
plt.hist(Y_s,20,(-1,1), histtype='bar',color='g')
plt.show()
```

```python
X_s=['A', 'B', 'C', 'D']
Y_s= [40, 30, 20, 10]
plt.title('Круговая диаграмма')
plt.pie(Y_s,labels=X_s,colors=['g','m','k','0.8'])
plt.show()
```
